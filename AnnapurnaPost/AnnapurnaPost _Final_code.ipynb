```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import urllib3
from bs4 import BeautifulSoup as BS
import re
from time import sleep
from random import randint

http = urllib3.PoolManager()
http.addheaders = [('User-agent', 'mozilla/61.0')]

Title=[]
Content=[]
Category=[]
Date=[]
Author_url=[]
Author_name=[]
URL=[]
Discription=[]
base_url = 'https://annapurnapost.com'

from google.colab import drive
drive.mount('/content/drive', force_remount=True)


#define a url, which wll be scrapped
for i in range(449000,450000):  ##Chnage range as you need
  url = 'https://annapurnapost.com/story/' + str(i)
  URL.append(url)
  print(url)

  sleep(randint(2,10))
  news_page = requests.get(url)
  news_soup = BS(news_page.content, 'html5lib')

  # extract title of the news
  # Use find() to get only the first h1 element
  title_element = news_soup.find('h1')
  if title_element:
    title = title_element.text.strip() # Strip whitespace for consistency
    Title.append(title)
    print(title)
  else:
    Title.append(None) # Append None if no title is found
  #Discription
  Discription_element = news_soup.find('meta', attrs={'name': 'description'})
  if Discription_element:
    Discription.append(Discription_element.get('content'))
    print(Discription_element.get('content'))
  else:
    Discription.append(None)

  # find the news content
  content_element = news_soup.select_one("div.news__details")
  if content_element:
    news_content = ""
    for content in content_element.findAll("p"):
      news_content += content.text.strip() + " "
    Content.append(news_content.strip())
    print(news_content.strip())
  else:
    Content.append(None)

  #extract news category
  category_element = news_soup.select_one(".card__category")
  if category_element:
    category = category_element.text.strip()
    Category.append(category)
    print(category)
  else:
    Category.append(None)

  #extract time and date
  time_element = news_soup.select_one(".date")
  if time_element:
    time = time_element.text.strip()
    Date.append(time)
    print(time)
  else:
    Date.append(None)

  #extract author_url and author_name
  author_element = news_soup.select_one(".author__name")
  if author_element:
    author_url =  base_url + author_element.a.get('href')
    author_name = author_element.text.strip()
    Author_url.append(author_url)
    Author_name.append(author_name)
    print(author_url)
    print(author_name)
  else:
    Author_url.append(None)
    Author_name.append(None)



df=pd.DataFrame({'URL':URL,'Title':Title,'Discription':Discription,'Content':Content,'Category':Category,'Date':Date,'Author_url':Author_url,'Author_name':Author_name})
df.to_csv('/content/drive/MyDrive/AnnapurnaPost/annapurnapost11.csv')

# df=pd.DataFrame({'URL':URL,'Title':Title,'Discription':Discription,'Content':Content,'Category':Category,'Date':Date,'Author_url':Author_url,'Author_name':Author_name})
# df.to_csv('annapurnapost.csv')
```
