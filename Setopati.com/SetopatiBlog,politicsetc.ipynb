```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
from time import sleep
import time
import random
URL = []
Title = []
Category = []
Author_name = []
Author_url = []
Date = []
Content = []
Description = []

# Base URL for constructing full URLs
base_url = 'https://www.setopati.com'

# List of sections to scrape
sections = ['politics', 'opinion', 'social','blog']

# Function to get the full content of an article
def get_article_details(article_url):
    response = requests.get(article_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract author details
        author_box = soup.find('h2', class_='main-title')
        if author_box:
            author_name = author_box.find('a').get_text(strip=True)
            author_url = author_box.find('a')['href']
            Author_name.append(author_name)
            Author_url.append(author_url)
        else:
            author_name = 'Unknown'
            author_url = ''
            Author_name.append(author_name)
            Author_url.append(author_url)

        # Extract date and time
        date_time_element = soup.find('span', class_='left-icon fa fa-clock-o')
        if date_time_element:
            date_time = date_time_element.get_text(strip=True)
            Date.append(date_time)
        else:
            date_time = 'Date not found'  # Handle the case where date is not found
            Date.append(date_time)

        # Extract description
        description_tag = soup.find('span', class_='maintitle')
        description = description_tag.get_text(strip=True) if description_tag else 'No description available'
        # Description.append(description)

        # Find the content within <p style="text-align: justify;"> sections
        paragraphs = soup.find_all('p', style='text-align: justify;')
        content = '\n'.join([p.get_text(strip=True) for p in paragraphs])
        Content.append(content)

        return author_name, author_url, date_time, description, content
    else:
        return 'Failed to retrieve content', '', '', '', ''


# Function to scrape a specific section and page
def scrape_section(section, page):
    url = f'{base_url}/{section}?page={page}'
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all article items
        article_items = soup.find_all('div', class_=['row bishesh news-cat-list','row bishesh news-cat-list alt','items col-md-4','items col-md-6'])

        for article in article_items:
            # Extract the title and title link
            a_tag = article.find('a', href=True)
            if a_tag:
                title = a_tag.get('title', '').strip()
                title_link = a_tag['href']
                Title.append(title)
                URL.append(title_link)
                # Extract the category from the URL
                category = title_link.split("/")[-2]
                category
                Category.append(category)
                #descrepton
                discription = None
                Description.append(discription)






                # Construct the full URL if necessary
                full_url = f'{base_url}{title_link}' if title_link.startswith('/') else title_link
                # URL.append(full_url)

                # Get additional details using the function
                author_name, author_url, date_time, description, content = get_article_details(full_url)

                print(f"Title: {title}")
                print(f"Title Link: {full_url}")
                print(f"Category: {category}")
                print(f"Author Name: {author_name}")
                print(f"Author URL: {author_url}")
                print(f"Date and Time: {date_time}")  # Now handles potential missing date
                print(f"Description: {description}")
                print(f"Content:\n{content}\n")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

# Loop through each section and each page
for section in sections:
    for page in range(1, 2):  # Adjust range as needed to cover more pages
        print(f"Scraping {section}, page {page}")
        scrape_section(section, page)
        time.sleep(random.randint(2, 10))  # Sleep for a random time between 2 and 10 seconds
#mount in google drive

  # Convert lists to DataFrames
df = pd.DataFrame({'URL': URL, 'Title': Title, 'Description':Description,'Content': Content,'Category': Category,'Date': Date,  'Author_url': Author_url,'Author_name': Author_name  })
df.to_csv('setopat.csv', index=False)

```
